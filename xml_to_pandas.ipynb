# Converter XML para dataframe Pandas

Importar as biblitecas necessárias

```python


import xml.etree.ElementTree as ET
import pandas as pd
from xmltopd import extract_dataframe
```

Ler o arquivo e destrinchar o xml

```python
doc_list: list[dict] = extract_dataframe("globo.xml")
doc_list
```

Criar um *dataframe* Pandas a partir dos dados

```python
news_df: pd.DataFrame = pd.DataFrame(doc_list)[["title", "description", "category"]]
news_df
    
```
