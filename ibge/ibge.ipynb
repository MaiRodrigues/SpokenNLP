Web-scrapping

```python
with open("ibge.html", 'r', encoding="utf-8") as f:
    file_contents = f.read()
    
soup = BeautifulSoup(file_contents, "html.parser")
text_body_not_stilized = soup.find(name="a")

text_body_not_stilized.attrs

# text_body = BeautifulSoup(text_body_not_stilized.string, "html.parser").find(name="div", class_="texto--single")

# children = text_body.findChildren("p")
```

```python
from bs4 import Tag

html_array: list[str] = []

for item in items:
    text: str = ""
    html: str = requests.get(item.get("link", "")).text
    text_paragraphs: list[Tag] = BeautifulSoup(html, "html.parser").find(name="div", class_="texto--single").findChildren("p")

    for p in text_paragraphs:
        text = text + p.text
    html_array.append(text)
    
html_array
```

Cria Pandas dataframe

```python
dataframe: pd.DataFrame = pd.DataFrame(items)
dataframe["texto"] = html_array
dataframe
```
